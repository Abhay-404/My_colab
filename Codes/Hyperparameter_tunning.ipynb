```
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import lightgbm as lgb


from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import lightgbm as lgb
import numpy as np

def train_and_evaluate_cv(X_train, y_train, params, n_splits=5):
    """
    Trains an LGBM model with the given parameters using Stratified K-Fold cross-validation
    and evaluates its performance over the cross-validation folds.

    Args:
        X_train (pd.DataFrame): Training features.
        y_train (pd.Series): Training target.
        params (dict): Dictionary of LGBM model parameters.
        n_splits (int): Number of cross-validation splits. Default is 5.

    Returns:
        dict: A dictionary containing the average accuracy, precision, recall, F1-score, and ROC AUC
              over the cross-validation folds.
    """
    # Initialize stratified k-fold
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

    # Initialize lists to store metrics for each fold
    accuracies, precisions, recalls, f1_scores, roc_aucs = [], [], [], [], []

    # Set LGBM to use GPU if available
    params['device'] = 'gpu'   #  if gpu hai nhi to comment krdo
    params['verbose'] = -1

    # Get imbalance ratio and set scale_pos_weight
    imbalance_ratio = y_train.value_counts()[1] / y_train.value_counts()[0]
    params['scale_pos_weight'] = imbalance_ratio

    # Loop over each fold
    for train_idx, val_idx in skf.split(X_train, y_train):
        X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]

        # Create LGBM datasets
        train_data = lgb.Dataset(X_train_fold, label=y_train_fold)
        val_data = lgb.Dataset(X_val_fold, label=y_val_fold)

        # Train LGBM model
        model = lgb.train(params,
                          train_data,
                          valid_sets=[train_data, val_data],
                          callbacks=[lgb.early_stopping(stopping_rounds=100)])

        # Make predictions on validation set
        y_pred = model.predict(X_val_fold)
        y_pred_binary = (y_pred > 0.5).astype(int)  # Convert probabilities to binary predictions

        # Calculate metrics for this fold
        accuracies.append(accuracy_score(y_val_fold, y_pred_binary))
        precisions.append(precision_score(y_val_fold, y_pred_binary))
        recalls.append(recall_score(y_val_fold, y_pred_binary))
        f1_scores.append(f1_score(y_val_fold, y_pred_binary))
        roc_aucs.append(roc_auc_score(y_val_fold, y_pred))
        print(f1_scores)

    # Return average metrics across all folds
    return {
        'accuracy': np.mean(accuracies),
        'precision': np.mean(precisions),
        'recall': np.mean(recalls),
        'f1_score': np.mean(f1_scores),
        'roc_auc': np.mean(roc_aucs)
    }


```
