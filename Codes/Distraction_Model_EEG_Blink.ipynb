```python
def calculate_statistics(df_chunk):
    stats = {}
    # Exclude the 'result' column from statistical calculations
    for column in df_chunk.columns:
        if column != 'result':
            stats[f'{column}_mean'] = df_chunk[column].mean()
            stats[f'{column}_median'] = df_chunk[column].median()
            stats[f'{column}_std'] = df_chunk[column].std()
            stats[f'{column}_var'] = df_chunk[column].var()
            stats[f'{column}_min'] = df_chunk[column].min()
            stats[f'{column}_max'] = df_chunk[column].max()
            stats[f'{column}_range'] = df_chunk[column].max() - df_chunk[column].min()
    return stats

# Split the data into chunks of 3 rows each
chunks = [df_blink_filtered.iloc[i:i+3] for i in range(0, len(df_blink_filtered), 3)]

result_df = pd.DataFrame()

# Calculate statistics for each chunk and append to the result DataFrame
for i, chunk in enumerate(chunks):
    # Check if all rows in the chunk have the same 'result' value
    if chunk['result'].nunique() == 1:
        stats = calculate_statistics(chunk)
        stats['result'] = chunk['result'].iloc[0]  # Add the consistent 'result' value
        stats_df = pd.DataFrame([stats])  # Convert the stats dictionary to a DataFrame
        stats_df['chunk'] = i + 1  # Add a chunk identifier
        result_df = pd.concat([result_df, stats_df], ignore_index=True)

```
