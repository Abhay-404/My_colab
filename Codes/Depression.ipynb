### Adding statistical feature into the df

```python
def calculate_statistics(df_chunk):
    stats = {}
    # Exclude the 'result' column from statistical calculations
    for column in df_chunk.columns:
        if column != 'result':
            stats[f'{column}_mean'] = df_chunk[column].mean()
            stats[f'{column}_median'] = df_chunk[column].median()
            stats[f'{column}_std'] = df_chunk[column].std()
            stats[f'{column}_var'] = df_chunk[column].var()
            stats[f'{column}_min'] = df_chunk[column].min()
            stats[f'{column}_max'] = df_chunk[column].max()
            stats[f'{column}_range'] = df_chunk[column].max() - df_chunk[column].min()
    return stats

# Filter the data based on 'result' column for 0 and 1 separately
df_result_0 = df_blink_filtered[df_blink_filtered['result'] == 0]
df_result_1 = df_blink_filtered[df_blink_filtered['result'] == 1]

# Split the filtered data into chunks of 3 rows each
chunks_0 = [df_result_0.iloc[i:i+3] for i in range(0, len(df_result_0), 3)]
chunks_1 = [df_result_1.iloc[i:i+3] for i in range(0, len(df_result_1), 3)]

result_df = pd.DataFrame()

# Function to process chunks and append statistics
def process_chunks(chunks, result_value):
    global result_df
    for i, chunk in enumerate(chunks):
        if len(chunk) == 3 :  # Ensure the chunk has exactly 3 rows
            stats = calculate_statistics(chunk)
            stats['result'] = result_value  # Add the consistent 'result' value
            stats_df = pd.DataFrame([stats])  # Convert the stats dictionary to a DataFrame
            stats_df['chunk'] = i + 1  # Add a chunk identifier
            result_df = pd.concat([result_df, stats_df], ignore_index=True)

# Process chunks for result == 0
process_chunks(chunks_0, 0)

# Process chunks for result == 1
process_chunks(chunks_1, 1)
```
